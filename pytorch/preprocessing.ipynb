```
import numpy as np
import pandas as pd
pd.options.display.max_rows = 100
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
train_labels_df = pd.read_csv('/content/drive/My Drive/Practice/FOODC/train.csv')
print(train_labels_df.head(2))
print(train_labels_df.info())
```

```
train_labels_df['ClassName'] = train_labels_df['ClassName'].astype('category')
train_labels_df.head(2)
```

```
train_labels_df['ClassName_cat'] = train_labels_df['ClassName'].cat.codes
print(train_labels_df.head(2))
print("CATEGORIES: ",train_labels_df['ClassName_cat'].nunique())
```

```
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'water'").sample(frac=.61).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'bread-white'").sample(frac=.49).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'salad-leaf-salad-green'").sample(frac=.45).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'bread-wholemeal'").sample(frac=.25).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'coffee-with-caffeine'").sample(frac=.18).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'tomato'").sample(frac=.17).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'butter'").sample(frac=.15).index)
```

```
times_2 = ["egg", "jam", "espresso-with-caffeine", "apple", "banana", "tea", "mixed-salad-chopped-without-sauce" ,"potatoes-steamed", "wine-white", "cheese", 
           "chips-french-fries", "white-coffee-with-caffeine", "hard-cheese", "tomato-sauce", "pizza-margherita-baked", "cucumber", "chicken"]

times_3 = ["pasta-spaghetti", "zucchini", "dark-chocolate", "avocado", "leaf-spinach", "strawberries", "water-mineral", "boisson-au-glucose-50g", "sweet-pepper",
           "broccoli", "mixed-nuts", "beer", "french-beans", "croissant", "bread-whole-wheat", "gruyere", "tea-green", "almonds", "mandarine", "biscuits", "parmesan",
           "ham-raw", "honey", "bread-sourdough", "sauce-savoury", "bread-french-white-flour", "salmon", "salami", "soft-cheese", "mayonnaise", "onion", "corn", "pickle"]

reps = [2 if val in times_2 else 3 if val in times_3 else 1 for val in train_labels_df.ClassName]
train_labels_df = train_labels_df.loc[np.repeat(train_labels_df.index.values, reps)]
```

```
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'egg'").sample(frac=.24).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'espresso-with-caffeine'").sample(frac=.24).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'apple'").sample(frac=.24).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'banana'").sample(frac=.24).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'tea'").sample(frac=.24).index)
train_labels_df = train_labels_df.drop(train_labels_df.query("ClassName == 'mixed-salad-chopped-without-sauce'").sample(frac=.24).index)
```

```
train_labels_df = train_labels_df.sample(frac=1) #SHUFFLE RANDOMLY ALL ROWS
```
